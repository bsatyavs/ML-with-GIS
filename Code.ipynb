```python
import geopandas as gpd
```

```python
california_housing = gpd.read_file("california_housing.shp")
```

```python
california_housing.head(10)
```

```python
# task 1 :

# getting x and y values from the geometry column

california_housing['x'] = california_housing['geometry'].x
california_housing['y'] = california_housing['geometry'].y
california_housing.head()
```

```python
df = california_housing.drop('geometry', axis=1)
df.head(10)
```

```python
# we have a column ocean_prox which is categorical, I am converting it into numeric 

# I am using one hot encoding to convert them into numeric
import pandas as pd

df = pd.get_dummies(df, columns=['ocean_prox'])
```

```python
# TASK 2 implementing Random Forests
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
x = df.drop('median_val', axis=1)  # putting all the columns except the target variable in x
y = df['median_val']               # putting the Target variable in y
```

```python
# splitting data set into train and test
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# TASK 2
model = RandomForestRegressor(random_state=42, min_samples_leaf=10)
```

```python
model.fit(X_train, y_train)
```

```python
# Making predictions on the test set
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
# TASK 3

from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np

rmse_scorer = make_scorer(lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true, y_pred)), greater_is_better=False)

# Setting up 10-fold cross-validation
kfold = KFold(n_splits=10, shuffle=True, random_state=42)

# Performing the cross-validation
rmse_scores = cross_val_score(model, x, y, cv=kfold, scoring=rmse_scorer)

# Calculating the average RMSE across all 10 folds
average_rmse = np.abs(np.mean(rmse_scores))

print(f"Average RMSE: {average_rmse}")
```

```python
# TASK 4

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [80, 100, 120]
}

# Setting up GridSearchCV
grid_search = GridSearchCV(estimator = model, param_grid=param_grid, cv=10, scoring='neg_root_mean_squared_error', verbose=1)

# Fitting GridSearchCV
grid_search.fit(X_train, y_train)

# Gettgin the best estimator
best_rf = grid_search.best_estimator_

# Evaluating on the test data
y_pred = best_rf.predict(X_test)
best_rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Printing the results
print("Best parameters:", grid_search.best_params_)
print("Best cross-validation RMSE: ", -grid_search.best_score_)
print("Test set RMSE: ", best_rmse)
```
